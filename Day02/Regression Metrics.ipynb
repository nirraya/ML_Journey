```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('income_happiness.csv')
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset.shape
```

```python
plt.scatter(dataset['income'],dataset['happiness'])
```

```python
X=dataset['income'].values #income per 10 thousand dollars
```

```python
y=dataset['happiness'].values #in a range of 0-10
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
X_train.shape
```

```python
X_train = X_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
```

```python
lr.fit(X_train, y_train)
```

```python


# predict the y values for each x value in X_test
y_pred = lr.predict(X_test)

# plot the scatter plot
plt.scatter(X, y)

# plot the line of best fit
plt.plot(X_test, y_pred, color='red')

# add x label, y label and title
plt.xlabel('Income per 10 Thousand USD')
plt.ylabel('Happiness Score in range 0-10')
plt.title('Hapiness as a Function of Income')

plt.show()
```

```python
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(14,6))

# plotting the scatter plot
ax1.scatter(X,y)

# adding x label, y label and title
ax1.set_xlabel('Income per 10 Thousand USD',fontsize=15)
ax1.set_ylabel('Happiness Score in Range 0-10',fontsize=15)


# plotting the scatter plot
ax2.scatter(X, y)
# plotting the line of best fit
ax2.plot(X_test, y_pred, color='red')

# adding x label, y label
ax2.set_xlabel('Income per 10 Thousand USD',fontsize=15)
ax2.set_ylabel('Happiness Score in Range 0-10',fontsize=15)

#adding the figure title
fig.suptitle('Hapiness as a Function of Income',fontsize=20)

plt.show()
```

```python
# To calculate matrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
y_pred = lr.predict(X_test)
```

```python
print(f'MAE: {mean_absolute_error(y_test,y_pred)}')
```

```python
r2=r2_score(y_test,y_pred)
print(f'R2 Score: {r2}')
```

```python
n=X_test.shape[0]
k=1 #no. of independent columns
```

```python
#adjusted r2 score
adj_r2= 1-((1-r2)*(1-n)/(n-1-k))
print(adj_r2)
```

# Adding random column to check the changes in r2 score. r2 score will increase but adjusted r2 score will decrease.

```python
new_dataset=dataset.copy()
new_dataset['random']=np.random.random(498)
```

```python
new_dataset.head()
```

```python
plt.scatter(new_dataset['random'],new_dataset['happiness'])
plt.xlabel("Random")
plt.ylabel("Hapiness Score in Range 0-10")
```

```python
X=new_dataset.iloc[:, [0, 2]].values
y=new_dataset['happiness'].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
r2=r2_score(y_test,y_pred)
print(f'R2 Score: {r2}')
```

```python
n=X_test.shape[0]
k=2 #no. of independent columns
```

```python
#adjusted r2 score
adj_r2= 1-((1-r2)*(1-n)/(n-1-k))
print(adj_r2)
```
